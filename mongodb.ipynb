Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
MongoDB over SQL databases?

 MongoDB is a document-oriented NoSQL database.
    NoSQL databases are a type of database that do not follow the traditional relational database model.
    Instead, they store data in different ways, such as in JSON documents, key-value pairs, or graphs.
    
    
    MongoDB stores data in JSON-like documents, which are essentially key-value pairs with additional properties.
        This makes it a good choice for storing unstructured or semi-structured data.
        
        
        
         advantages of using MongoDB over SQL databases:
                
         1. Flexibility:-MongoDB's document model is very flexible, which makes it a good choice for
                       storing data that does not fit neatly into a traditional relational schema.      
                
          2.Scalability:-MongoDB is highly scalable, both horizontally and vertically.
            This means that you can easily add more data to your database without having
            to worry about performance issues. 
            
         3.Performance:-MongoDB is generally faster than SQL databases for certain types of queries,
            such as those that involve joins.

Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to
demonstrate this.

      find() method in MongoDB is used to query a database collection and retrieve documents that match certain criteria. It           allows you to specify filters, sorting options, projection of specific fields,
      
      
      
      find() method, you call it on a collection object and provide a query as a parameter.
      The query is a document that specifies the criteria for matching documents.
      
      
      
      Filtering: You can specify conditions to filter documents based on specific field
      values using comparison operators like $eq, $ne, $gt, $lt, $gte, $lte, and logical 
      operators like $and, $or, $not, etc.
      
      
      Projection: You can specify which fields to include or exclude in the returned documents
      by providing a projection document using the projection parameter. 
      
      
      Sorting: You can sort the returned documents based on specific fields using the sort parameter
